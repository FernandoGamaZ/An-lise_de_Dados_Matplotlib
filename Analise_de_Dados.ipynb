```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_excel("dados_vendas.xlsx")
display(df)
```

```python

#Quantidade de vendas diarias

plt.figure(figsize=(20,5))
df['Quantidade Vendida'].plot(marker = 'o', linestyle = '-', color= 'red')
plt.title("Vendas Diárias")
plt.xlabel("Data")
plt.ylabel("Quantidade Vendida")
plt.grid()
plt.show()
```

```python
df_max = df['Quantidade Vendida'].max()
df_min = df['Quantidade Vendida'].min()
media_vendas = df["Quantidade Vendida"].mean()

print(f"O máximo de vendas diárias foi {df_max}")
print('-----------------------------------------------')

print(f"O mínimo de vendas diárias foi {df_min}")
print('-----------------------------------------------')
print(f"A média de vendas diárias foi {media_vendas}")

print(df_min)
```

```python
df["Data"] = pd.to_datetime(df["Data"])
df = df.set_index("Data") # Set 'Data' as the index and update df
Mes_vendas = df["Quantidade Vendida"].resample("ME").sum() # Resample by month-end and sum monthly sales

# 'Data' is already the index, so we can directly resample for daily sales
dia_vendas = df["Quantidade Vendida"].resample("D").sum()
```

```python
plt.figure(figsize=(20,5))
dia_vendas.plot(marker = 'o', linestyle = '-', color = 'purple')
plt.axhline(y=df_max, color = 'green', linestyle = '--', linewidth = 2, label=f"Maximo: {df_max}")
plt.axhline(y = df_min, color = 'red', linestyle = '--', linewidth = 2, label=f"Minimo:{df_min}")
plt.axhline(y = media_vendas, color = 'blue', linestyle = '--', linewidth = 2, label= f"Média: {media_vendas}")
plt.legend()
plt.title("Quantidade de Vendas Diárias")
plt.xlabel("Dia")
plt.ylabel("Quantidade Vendida")
plt.grid()
```

```python
media_mes = Mes_vendas.mean()
min_mes = Mes_vendas.min()
max_mes = Mes_vendas.max()

print(f"O máximo de vendas por mês foi de {max_mes}")
print('-----------------------------------------------')
print(f"O minimo de vendas por mês foi de {min_mes}")
print('-----------------------------------------------')
print(f"A média de vendas por mês foi de {media_mes}")
```

```python
plt.figure(figsize=(30,10))
Mes_vendas.plot(kind = 'bar', color = 'red')
plt.axhline(y = max_mes, color = 'blue', linestyle = '--', label = f"Máximo: {max_mes}")
plt.axhline(y = min_mes, color = 'brown', linestyle = '--', label= f"Mínimo: {min_mes}")
plt.axhline(y = media_mes, color = 'green', linestyle = '--', label=f"Média: {media_mes}")
plt.legend(loc = 'best', fontsize = 12)
plt.title("Vendas por Mês")
plt.xlabel("Mês")
plt.ylabel("Quantidade de vendas")
plt.xticks(rotation=0)
plt.grid()
```
